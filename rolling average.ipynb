```python
from nsepy import *
from datetime import *
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sym=["SBIN","BHEL","ACC","GAIL","TATAMOTORS","TCS","INFY","SUNPHARMA","IDEA","RELIANCE"]
```

```python
d= pd.DataFrame()
j=0
for i in sym:
    result = get_history(symbol=i, start=date(2015,1,1), end=date(2015,12,31))
    s2=result[['Close']]
    s2=s2.rename(columns = {'Close':i})
    if j==0:
        d=d.append(s2)
        j=j+1
    else:
        d=d.join(s2)
```

```python
d1=d.reset_index()
```

```python
d1['Date'] = d1['Date'].apply(pd._libs.tslib.normalize_date)
```

```python
d1['Week'] = d1['Date'].dt.week
```

```python
d1=d1[d1['Week']%4==0]
```

```python
d2=d1.drop(["SBIN","BHEL","ACC","GAIL","TATAMOTORS","TCS","INFY","SUNPHARMA","IDEA","RELIANCE"], axis=1)
```

```python
for i in sym:
    d2['MA-'+i]=d1[i].rolling(2).mean()
```
