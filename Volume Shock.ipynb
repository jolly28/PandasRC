```python
import numpy as np
from nsepy import *
from datetime import *
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
result = get_history(symbol='SBIN', start=date(2015,1,1), end=date(2015,12,31))
```

```python
result=result.reset_index()
r1=result
r1["V"]=np.nan
```

```python
for i in range(0,247):
    p=((r1.Volume[i+1]-r1.Volume[i])/(r1.Volume[i]))*100
    if p>=10:
        r1.V[i]=10
    elif p<=-10:
        r1.V[i]=-10
    else:
        r1.V[i]=0
```

```python
r1['Volume Shocks']=result.V.map({10:0,-10:1,0:'NaN'})
r1=r1.drop(['V'],axis=1)
```
