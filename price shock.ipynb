```python
import numpy as np
from nsepy import *
from datetime import *
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
result = get_history(symbol='SBIN', start=date(2015,1,1), end=date(2015,12,31))
```

```python
result=result.reset_index()
r1=result
r1["P"]=np.nan
```

```python
r1
```

```python
for i in range(0,247):
    p=((r1.Close[i+1]-r1.Close[i])/(r1.Close[i]))*100
    if p>=2:
        r1.P[i]=2
    else:
        r1.P[i]=0
```

```python
r1['Price Shocks']=result.P.map({2:1,0:0})
r1=r1.drop(['P'],axis=1)
```
